```python
with open("the-verdict.txt","r",encoding="utf-8") as f:
    raw_text = f.read()
```

```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
enc_text = tokenizer.encode(raw_text)
```

```python
print(len(enc_text))
```

```python
enc_sample  = enc_text[50:]
# enc_sample
```

```python
#context size tells us how many words you give to predict next words e.g. if context size is four then you give four words and the model predicts the fifth word. e.g apple is a tastythen it will predict fruit
```

```python
#for now let us take context size as 4
context_size=4
x=enc_sample[:context_size]
y=enc_sample[1:context_size+1] #but here y should be just one token na? means context_size:contet_size+1

print(f"x: {x}")
print(f"y:      {y}")
```

```python
for i in range(1,context_size+1):
    context = enc_sample[:i]
    desired = enc_sample[i]

    print(context,"------>",desired)
```

```python
for i in range(1,context_size+1):
    context = enc_sample[:i]
    desired = enc_sample[i]

    print(tokenizer.decode(context), "------>" , tokenizer.decode([desired]))
```
